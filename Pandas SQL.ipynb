* `pandasql` allows to query pandas DataFrame using SQL syntax. 

```python
import pandas as pd
import pandasql as psql
```

```python
df = pd.read_csv('data/aadhaar_data.csv')
```

```python
df.head()
```

```python
df.rename(columns= lambda x: x.replace(' ', '_').lower(), inplace = True)
```

```python
df.shape
```

* Let's choose first 20 rows and few columns from dataframe using pandasql

```python
q = 'SELECT Registrar, Enrolment_Agency FROM df LIMIT 20;'
```

```python
psql.sqldf(q, locals())  # locals() and globals() showing session/env variables
```

```python
q = "SELECT * FROM df WHERE state='Gujarat';"
```

```python
psql.sqldf(q, locals())
```

```python
q = 'SELECT district, sum(aadhaar_generated) FROM df GROUP BY district'
```

```python
psql.sqldf(q, locals())
```

```python
q = 'SELECT district, sub_district, sum(aadhaar_generated) FROM df GROUP BY district, sub_district'
```

```python
psql.sqldf(q, locals())
```

```python
q = 'SELECT district, sub_district, sum(aadhaar_generated) FROM df WHERE age > 60 GROUP BY district, sub_district'
```

```python
psql.sqldf(q, locals())
```
